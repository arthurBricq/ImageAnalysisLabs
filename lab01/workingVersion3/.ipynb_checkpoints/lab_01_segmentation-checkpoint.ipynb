## Part 1: Brain segmentation
In this section, we have a picture of the brain obtained after an fMRI. We want to to count the number of pixels included in the brain. In order to do so, we are going to present 3 different methods, followed by a recap of the results obtained.

### 1.1 Brain image visualization

```python
# Load image
brain_im = skimage.io.imread('data/lab-01-data/brain-slice40.tiff')
im_h, im_w = brain_im.shape

# Plot the image
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.imshow(brain_im, cmap='gray')
ax.set_title('MRI brain image ({} px, {} px)'.format(im_h, im_w))
ax.axis('off')
plt.show()
```

### 1.2 Region growing

Region growing is an easy way to count the number of pixels within one region of the picture, here the brain. The region is defined by connectivity and homogeneity. Here we used 4-connectivity and tried 2 methods for homogeneity.
The first one uses static homogeneity : starting from 1 pixel, the region is constructed by 4-connectivity and small difference in pixel value with the value of the first chosen pixel.

The second method uses dynamic homogeneity : starting from 1 pixel, the region is also constructed by 4-connectivity and small difference in pixel value with its neighbours.

Both of them give convincing results.

```python
start_coord=(115,150)
regionSize = imPro.regionGrowing(brain_im,start_coord=(115,150),maxdiff=25,dynamic_homogeneity=False)
print ("Pixel value : ",brain_im[start_coord])
```

### 1.3 Contour detection

We can use contour detection to count the number of pixels within the brain in the picture. There must be an algorithm that finds the contour in the picture, followed by another algorithm that counts the number of pixels in a delimited region. In order to have reusable functions (for part II) we wrote a function that takes as input the edge map (obtained after contour detection) and will cluster every pixel within a closed group delimited by the boundary.

As we wanted to test the robustness of our results, we also programmed 2 different methods to find contours. One of them is using the Lagrange technique, as the other one is an heuristic way to find edges. 

#### Different methods to find edges:

```python
# 1) Laplace Method
def getEdgeMapMethod1(img, t1, t2):
    binary_im = np.logical_and(img > t1, brain_im < t2) 
    edgeMap = skimage.filters.laplace(binary_im)
    edgeMap[edgeMap>0]=-edgeMap[edgeMap>0]
    return edgeMap
   
# 2) Heuristic method
def getEdgeMapMethod2(img, t1, t2):
    sigma=2.5
    G_t1=10
    G_t2=200
    (edges,thld_img,G,Phi,G_thld_img)=imPro.skeletonize_based_sobel(img,sigma,t1,t2,G_t1,G_t2)
    edgeMap = -np.array(edges).astype(int)
    return edgeMap

# 3) Canny Edge Detector
def getEdgeMapMethod3(img):
    edgeMap = canny(img, sigma= 1.1)
    edgeMap = skimage.morphology.dilation(edgeMap)
    edgeMap = -edgeMap.astype(int)
    return edgeMap
```

#### Illustration of the three edge detector that we used

```python
fig,axs = plt.subplots(1,3,figsize=(17,10))
for method,ax in zip(range(3),axs):
    if method == 0: edgeMap = getEdgeMapMethod1(brain_im,40,100)
    elif method == 1: edgeMap = getEdgeMapMethod2(brain_im,55,105)
    elif method == 2: edgeMap = getEdgeMapMethod3(brain_im)
    ax.imshow(edgeMap,cmap='gray')
    ax.set_title('Method {}'.format(method+1))
```

The following block of code shows a method that is able to call our different edge detector and displays the results. 

```python
def segmentateBrain(brain_im, method):
    """
    This method will do the segmentation of the brain image.
    INPUTS:
    - brain_im: brain image in grayscale
    - method: integer that is either 1, 2 or 3
    """    
    ### 1. Get the edge map with method (1) or with method (2) or with method (3)
    if method == 1: edgeMap = getEdgeMapMethod1(brain_im,40,100)
    elif method == 2: edgeMap = getEdgeMapMethod2(brain_im,55,105)
    elif method == 3: edgeMap = getEdgeMapMethod3(brain_im)
    else: return False

    ### 2. Cluster every pixels within closed group with our algoribbthm
    labels = imPro.labelizePixels(edgeMap)

    ### 3. Analysis of the labels
    #plt.figure()
    # Histogram of each label
    n, bins = histogram(labels.ravel(),np.unique(labels))
    # Sort the occurences in ascending order
    indexes = np.argsort(n)
    # The label of interest is (clearly) the second most recurent label, since the most recurrent one is the background
    indexOfLabel = indexes[-2]
    
    ### 4. Display the results
    numberOfPixels = n[indexOfLabel] 
    ratio = numberOfPixels / im_h / im_w
    print("Result of the analysis for method",method)
    print("- Number of pixels of the brain: {},\n- Ratio: {}".format(numberOfPixels, ratio))
    print("")
    
```

#### Obtaining the results

```python
segmentateBrain(brain_im=brain_im, method = 1)
segmentateBrain(brain_im=brain_im, method = 2)
segmentateBrain(brain_im=brain_im, method = 3)
```

## Part 2: Shape/color segmentation

In this section, we create a routine to :

### 2.1 Visualization of the data

```python
# Load images
im1 = skimage.io.imread('data/lab-01-data/arena-shapes-01.png')
im2 = skimage.io.imread('data/lab-01-data/arena-shapes-02.png')
im3 = skimage.io.imread('data/lab-01-data/arena-shapes-03.png')
images = np.array([im1,im2,im3])
fig,axs = plt.subplots(1,3,figsize=(17,10))
for i,ax in enumerate(axs):
    ax.imshow(images[i])
```

### 2.2 Number of shapes of each color


This section is taking care of **counting the number of shapes of each color** for the three different pictures. 

The main assumption that we do is that each shape should contain approximatively the same number of pixels. Therefore, if we find a way to group each pixels within groups of closed forms (**segmentation**) and if we can do this for each color separatly, then we are done (and only some post-processing classification is required).

In the case of the given picture, we see that this assumption is verified (i.e. each shape has always the same size). There are 3 different shapes (*the line, the cross and the star*) that all have the same sizes and  are as follow : 

| Color | Line | Cross | Star |
| ----- | ---- | ----- | ---- |
| Blue | 0 | 3 | 2 |
| Black | 1 | 2 | 2 |

We want to find a robust way to construct this table from the pictures. Here, robust means that it should work with the 3 different pictures.

The code goes throug the following logical step for each picture:
1. Separate the background from all the shapes, using **optimal threshold** - so that it works for every intensity.
2. Obtain the black shapes (as logical np.array), using the blue chanel to threshold all the black shapes. For this, we use the knowledge of one picture that we know of being black, extract its blue value and extend it a little bit to perform the threshold. This is done with the function ` blueThresholdForPixel(pixel, image)`.
3. Obtain the blue shapes (as logical np.array), by basically doing a **logical substraction** of all the blue shapes with the black shapes. Since the contours would still be around, we then perform an **opening** to kill them. 
4. Count the number of pixels of each color. This parts makes use of the function `countNumberOfShapes(im_bi)`. The documentation of the function explains in the details how it works. Basically, it will call again the function `labelizePixels(edgeMap,minValue=0)` (as in part I: **segmentation**) to give a cluster to every pixel. It also uses the function `labels2Shapes(labels)` that does the post-processing job of counting every classifying every group of N pixels to a shape.

```python
for i,im in enumerate(images):
    # 1. Seperate the background from the shapes with optimal threshold 
    tmp = im[:,:,0]
    t = skimage.filters.threshold_otsu(tmp)
    im_bi = tmp < t 
    
    # 2. Obtain the black shapes
    tmp = im[:,:,2]
    tmp = skimage.filters.gaussian(tmp,sigma = 2.5) * 255
    tmp = skimage.exposure.rescale_intensity(tmp,out_range = (0,255))
    im_bi_black = imPro.blueThresholdForPixel((358,56),tmp)
    
    # 3. Obtain the blue shapes
    im_bi_blue = np.logical_xor(im_bi,im_bi_black)
    im_bi_blue = skimage.morphology.opening(im_bi_blue)
   
    # 4. Count the number of shapes for each picture
    groups_blue = imPro.countNumberOfShapes(im_bi_blue)
    groups_black = imPro.countNumberOfShapes(im_bi_black)
    
    # 5. Display the resuls 
    print("Image",i+1)
    print(" Color blue:")
    print("   Number of different elements with this color:", sum([len(row) for row in groups_blue]))
    print("   Number of different shapes :",len(groups_blue))
    print("   Description of each shapes: ")
    for group in groups_blue:
        print("     - Number of elements:", len(group)," ;  Mean number of pixel: ", np.mean(group))  
    print(" Color black:")
    print("   Number of different elements with this color:", sum([len(row) for row in groups_black]))
    print("   Number of different shapes :",len(groups_black))
    print("   Description of each shapes: ")
    for group in groups_black:
        print("   - Number of elements:", len(group)," ;  Mean number of pixel: ", np.mean(group))
    print("\n")
    
```

We can see that exactly the same results are obtained for the 3 images, and that those results are indeed corrects. The only counterpart of this code is that it is quite slow to run. 

### 2.3 Total area (in pixels) of each color

We want to count the number of pixels for each color. 

The first method that came to our mind was to apply the first same steps as before to dissociate the black and blue parts of the picture, and then we perform a simple counting of non-false elements. This method has the advantage of being very fast, as it only apply some binary thresholding on the picture and same filters.

```python
for i,im in enumerate(images):
    # 1. Seperate the background from the shapes with optimal threshold 
    tmp = im[:,:,0]
    t = skimage.filters.threshold_otsu(tmp)
    im_bi = tmp < t 

    # 2. Obtain the black shapes
    tmp = im[:,:,2]
    tmp = skimage.filters.gaussian(tmp,sigma = 2.5) * 255
    tmp = skimage.exposure.rescale_intensity(tmp,out_range = (0,255))
    im_bi_black = imPro.blueThresholdForPixel((358,56),tmp)
    
    # 3. Obtain the blue shapes
    im_bi_blue = np.logical_xor(im_bi,im_bi_black)
    im_bi_blue = skimage.morphology.opening(im_bi_blue)

    # Count the number of pixels on each color
    n_blue = np.count_nonzero(im_bi_blue)
    n_black = np.count_nonzero(im_bi_black)
    
    print("Image :",i+1)
    print("Color blue: {} pixels".format(n_blue))
    print("Color black: {} pixels".format(n_black))
    print("\n")
```

We observe that image 1 and image 3 have very similar results. However, the second image that has the brightest illumination tends to show less pixel. We therefore wanted to use a more robust method to see wether or not it was possible to obtain better result than this. It appeared that our different transformations (gaussian, intensity rescaling and Laplace edge detector) affected the shapes of each elements with a tendency to dilate the elements. It is possible to observe this by plotting the pre-processed image. 

In order to do so get ride of this fake 'dilation' of the shapes, we tried to apply a method that is slighty different and more difficult. It goes as following. For each picture:
1. Extract the edge map of the picture (including all the shapes) an perform a segmentation from this edge map. 
2. Extract the black parts and the blue parts of the picture (same step as before)
3. For each color, find the perform a segmentation (also like before) and find the center of each cluster of pixel
4. Then, look at each of the clusters' center in the first edgeMap and count the number of pixels there. 

Let us now introduce the code and the results obtained with this picture. 

```python
for i,im in enumerate(images):
    # 0. Seperate the background from the shapes with optimal threshold on the red channel
    tmp = im[:,:,0]
    t = skimage.filters.threshold_otsu(tmp)
    im_bi = tmp < t 

    # 1. Get the edge map (to be used later) and create
    edgeMap = skimage.filters.laplace(im_bi)
    edgeMap[edgeMap>0]=-edgeMap[edgeMap>0]
    labels = imPro.labelizePixels(edgeMap)

    # 2.a. Obtain the black shapes
    tmp = im[:,:,2]
    tmp = skimage.filters.gaussian(tmp,sigma = 2.5) * 255
    tmp = skimage.exposure.rescale_intensity(tmp,out_range = (0,255))
    im_bi_black = imPro.blueThresholdForPixel((358,56),tmp)

    # 2.b. Obtain the blue shapes
    im_bi_blue = np.logical_xor(im_bi,im_bi_black)
    im_bi_blue = skimage.morphology.opening(im_bi_blue)

    # 3. Find the centers of each cluster.
    centers_blue = imPro.findCentersOfClusters(im_bi_blue)
    centers_black = imPro.findCentersOfClusters(im_bi_black)

    # 4. From the clusters of each color, find the associated cluster in the edge map
    # and count the number of pixels within those clusters.
    pixels_blue = 0
    for center in centers_blue.astype(int):
        label = labels[center[0],center[1]]
        count = labels[labels==label].shape[0]
        pixels_blue += count

    pixels_black = 0
    for center in centers_black.astype(int):
        label = labels[center[0],center[1]]
        count = labels[labels==label].shape[0]
        pixels_black += count
    
    # 5. Display the results
    print("Image",i+1)
    print("   blue pixels: {}".format(pixels_blue))
    print("   black pixels: {}".format(pixels_black))
    print("\n")
```

We notice that the results obtained here are slighty lower than with the previous code, but the code takes a lot longer to run. Also, this code didn't fix the initial problem which was the fact that the second image has less pixels on every color. 

We can justifiy this observation by looking at the 3 pictures and by looking at the edgemap they generate. If one looks carefully, it is possible to see that the edges on the second image are slightly more compressed. This is because of the high brightness present on this picture.
